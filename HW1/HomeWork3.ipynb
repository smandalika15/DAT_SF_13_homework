```python
# Using get_dummies
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
from sklearn.cross_validation import cross_val_score
import numpy as np
bank = pd.read_csv("bank.csv", sep=";")
#Applying the rules below
#If a predictor only has two possible values, code it as a single binary variable (0 or 1). Since it's treated as a number, splits will naturally occur at 0.5.
#If a predictor has three or more possible values that are ordered, code it as a single variable (1, 2, 3, etc). Splits will naturally occur at 1.5, 2.5, etc.
#If a predictor has three or more possible values that are unordered, create dummy variables and drop one level as usual. The decision tree won't know that the dummy variables are related to one another, but that shouldn't matter in terms of predictive accuracy.
#If a predictor has thousands of possible unordered values, then it may be best to code it as a single variable (1, 2, 3, etc) instead of using dummy variables to minimize the size of the resulting model. (reference)

#Binary : Default, Housing, Loan 
#Apply dummy cols to: marital, contact

bank['default'] = bank.default.map({'no':0, 'yes':1})
bank['housing'] = bank.housing.map({'no':0, 'yes':1})
bank['y'] = bank.y.map({'no':0, 'yes':1})
bank['loan'] = bank.loan.map({'no':0, 'yes':1})
bank['job'] = bank.job.map({'admin.':0, 'blue-collar':1,'entrepreneur':2, 'housemaid':3,'management':4, 
                            'retired':5,'self-employed':6, 'services':7,'student':8,'technician':9, 'unemployed':10, 'unknown':11})
bank['month'] = bank.month.map({'jan':1, 'feb':2,'mar':3, 'apr':4,'may':5, 
                            'jun':6,'jul':7, 'aug':8,'sep':9,'oct':10, 'nov':11, 'dec':12})

bank['poutcome'] = bank.poutcome.map({'failure':0, 'other':1,'success':2, 'unknown':3 })
bank['education'] = bank.education.map({'primary':0, 'secondary':1,'tertiary':2, 'unknown':3 })
# create three dummy variables, drop the first dummy variable, and store this as a DataFrame
marital_dummies = pd.get_dummies(bank.marital, prefix='marital').ix[:, 1:]

contact_dummies = pd.get_dummies(bank.contact, prefix='contact').ix[:, 1:]

# concatenate the two dummy variable columns onto the original DataFrame
# note: axis=0 means rows, axis=1 means columns
bank = pd.concat([bank, marital_dummies,contact_dummies], axis=1)
feature_cols =['age', 'job', 'default', 'balance', 'housing', 'loan', 'day', 'duration', 'month',
 'campaign','pdays', 'previous', 'poutcome', 'marital_married', 'marital_single', 'contact_telephone', 'contact_unknown']

X = bank[feature_cols]
y = bank.y
# split into train/test
print X
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
# import class, instantiate estimator, fit with training set

treereg = DecisionTreeRegressor(max_depth=3,random_state=1)
treereg.fit(X_train, y_train)
preds = treereg.predict(X_test)
print np.sqrt(metrics.mean_squared_error(y_test, preds))
# print predictions and actual values
print preds
print y_test
```

```python
treeregcv = DecisionTreeRegressor(max_depth=3,random_state=1)
scores = cross_val_score(treeregcv, X, y, cv=5, scoring='mean_squared_error')
np.mean(np.sqrt(-scores))
print scores.mean()
# compute the "Gini importance" of each feature: the (normalized) total reduction of MSE brought by that feature
pd.DataFrame({'feature':feature_cols, 'importance':treereg.feature_importances_})
```
